```python
import json
import pandas as pd
import glob

files = glob.glob("../data/raw/*.json")
dfs = []
for file in files:
    with open(file, encoding="utf-8") as f:
        data = json.load(f)
        dfs.append(pd.json_normalize(data))
df = pd.concat(dfs, ignore_index=True)
```

```python
import pandas as pd
import sys
import os

# Thêm đường dẫn tuyệt đối tới thư mục src
sys.path.append(os.path.abspath("src"))

from cleaner import main_process



def get_dataframe(file_list):
    all_data = []
    for file in file_list:
        data = main_process(file)
        all_data.extend(data)
    
    df = pd.DataFrame(all_data)
    
    # Bung object 'address' thành các cột riêng biệt để EDA dễ hơn
    if 'address' in df.columns:
        address_cols = df['address'].apply(pd.Series)
        df = pd.concat([df.drop(columns=['address']), address_cols], axis=1)
        
    # Tương tự với 'spec' nếu cần
    if 'spec' in df.columns:
        spec_cols = df['spec'].apply(pd.Series)
        df = pd.concat([df.drop(columns=['spec']), spec_cols], axis=1)
        
    return df

# Chạy thử
files = ['../data/raw/rent_20251222.json', '../data/raw/bds_20251223.json']
df = get_dataframe(files)
print(df.info())
```
